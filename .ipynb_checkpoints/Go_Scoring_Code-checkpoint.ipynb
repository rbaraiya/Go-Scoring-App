```python
# Image Processing:
# Goal: Image -> 19x19 Array

class Board():
    
    def __init__(self, size, state, white_capped, black_capped): 
        self.size = size
        self.white_capped = white_capped
        self.black_capped = black_capped
        self.state = state # boardsize x boardsize array containing -1,1, and 0
    
    def is_valid(self):
        # a stone with no liberties cannot exist on the board.
        for i in range(self.size):
            for j in range(self.size):
                # stone can be in corner, edge
                pass
    def get_groups(self):
        # find all groups of connected stones on the board and return list of coordinates of the group with identity
        pass
        
            
    def get_liberties(self, i, j):
        # Return the number of liberties of a stone
        num_liberties = 0
        if self.state[i,j] == 0: # an empty spot has no property liberties
            return
        neighbors = self.get_neighbors(i,j)
        for identity in neighbors:
            if identity == 0:
                num_liberties += 1
        return num_liberties
            
    
    def get_neighbors(self, i, j):
        # Returns the identities (-1, 0, 1, or None) of neighbors of [i,j]
        
        if i == 0 and j != 0 and j != 8: # upper edge and not in the corners
                        
            left = self.state[i, j-1]
            right = self.state[i, j+1]
            top = None
            bottom = self.state[i+1, j]
            
        if i == 0 and j == 0: # top left corner
             
            left = None
            right = self.state[i, j+1]
            top = None
            bottom = self.state[i+1, j]
            
        if i == 0 and j == 8: # top right corner
            
            left = self.state[i, j-1]
            right = None
            top = None
            bottom = self.state[i+1, j]
            
        if i != 0 and i != 8 and j == 0: # left edge and not in the corners
            
            left = None
            right = self.state[i, j+1]
            top = self.state[i-1, j]
            bottom = self.state[i+1, j]
            
        if i == 8 and j == 0: # bottom left corner
            
            left = None
            right = self.state[i, j+1]
            top = self.state[i]
            bottom = None
            
        if i == 8 and j != 0 and j != 8: # bottom edge and not corners
            
            left = self.state[i, j-1]
            right = self.state[i, j+1]
            top = self.state[i-1, j]
            bottom = None
            
        if i == 8 and j == 8: # bottom right corner
            
            left = self.state[i, j-1]
            right = None
            top = self.state[i-1, j]
            bottom = None
        
        if i != 0 and i != 8 and j == 8: # right edge and not corners
            
            left = self.state[i, j-1]
            right = None
            top = self.state[i-1, j]
            bottom = self.state[i+1, j]
            
        elif i != 0 and i != 8 and j != 0 and j != 8: 
            
            left = self.state[i, j-1]
            right = self.state[i, j+1]
            top = self.state[i-1, j]
            bottom = self.state[i+1, j]
        
        return([left, right, top, bottom])
            
```
