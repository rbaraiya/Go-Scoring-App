```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('photos/test_go_board_1.jpg', cv2.IMREAD_GRAYSCALE)
print(img.shape)

plt.imshow(img)
```

```python
cropped_img = img[50:480, 150:650]
cv2.imshow("cropped", cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cropped_img = cv2.medianBlur(cropped_img,5)
cv2.imshow("blurred", cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Using Global Thresholding

threshold = 127
ret,th1 = cv2.threshold(cropped_img,threshold,255,cv2.THRESH_BINARY)
cv2.imshow(f"Global Thresholding (v = {threshold})", th1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Using Adaptive Thresholding

th2 = cv2.adaptiveThreshold(cropped_img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,19,4)
cv2.imshow("Adaptive Mean Thresholding", th2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
th3 = cv2.adaptiveThreshold(cropped_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,19,2)
cv2.imshow("Adaptive Gaussian Thresholding", th3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
